### formatage des quartier avec celle valide

```python
from fuzzywuzzy import process

# Liste des noms de quartiers corrects
quartiers_corrects = df_quartiers["Quartier"].tolist()

# Fonction pour corriger les noms de quartiers
def corriger_quartier(quartier):
    if pd.isna(quartier) or quartier.strip() == "":
        return "Non précisé"
    
    match, score = process.extractOne(quartier, quartiers_corrects)
    
    # Si le score de similarité est élevé (ex: > 80), on considère qu'il s'agit du bon quartier
    return match if score > 80 else "Non précisé"

# Appliquer la correction sur la colonne 'quartier_de_residence'
df_residence["quartier_de_residence"] = df_residence["quartier_de_residence"].apply(corriger_quartier)

# Afficher un aperçu des résultats
print(df_residence.head())

# Sauvegarder le fichier corrigé
df_residence.to_csv("df_residence_corrige.csv", index=False)
```

## formatage des arrondissements valide

```python
import re

# Dictionnaire de correspondance des chiffres arabes en chiffres romains
conversion_romaine = {
    "1": "I", "2": "II", "3": "III", "4": "IV", "5": "V", 
    "6": "VI", "7": "VII", "8": "VIII", "9": "IX", "10": "X"
}

# Fonction pour remplacer les chiffres par leur équivalent romain
def remplacer_par_romain(texte):
    match = re.search(r'\b(\d+)\b', texte)  # Trouver un chiffre seul
    if match:
        chiffre = match.group(1)  # Extraire le chiffre
        if chiffre in conversion_romaine:
            return texte.replace(chiffre, conversion_romaine[chiffre])  # Remplacer
    return texte  # Retourner inchangé si pas de correspondance

# Appliquer la transformation sur la colonne 'arrondissement_de_residence'
df_volontaire["arrondissement_de_residence"] = df_volontaire["arrondissement_de_residence"].apply(remplacer_par_romain)

# Vérification rapide
print(df_volontaire["arrondissement_de_residence"].unique())

# Sauvegarde du dataset corrigé
df_volontaire.to_csv("df_volontaire_clean_corrige.csv", index=False)

print("✅ Remplacement des chiffres par chiffres romains effectué avec succès !")
```

```python
# Liste des valeurs valides (Douala I, II, III... et "Douala (Non précisé)")
valeurs_valides = [f"Douala {romain}" for romain in ["I", "II", "III", "IV", "V", "VI"]] + ["Douala (Non précisé)"]

# Remplacer toutes les autres valeurs par "Non précisé"
df_volontaire["arrondissement_de_residence"] = df_volontaire["arrondissement_de_residence"].apply(
    lambda x: x if x in valeurs_valides else "Non précisé"
)

# Vérification rapide
print(df_volontaire["arrondissement_de_residence"].unique())

# Sauvegarde du dataset corrigé
df_volontaire.to_csv("df_volontaire_clean_corrige.csv", index=False)

print("✅ Nettoyage des arrondissements terminé avec succès !")
```

## Normalisation des religion

```python
df_volontaire = pd.read_csv("df_volontaire_clean_corrige.csv")
df_volontaire
```

```python
# Liste des religions (pour référence, mais pas nécessaire si elles sont déjà dans le DataFrame)
religions = [
    "Adventiste", "Animiste", "Aucune", "Baptiste", "CMCI", "Chrétien", 
    "Chrétien (Catholique)", "Chrétien (Non précisé)", "Chrétien (Né de nouveau)", 
    "Chrétien (Protestant)", "Chrétien (Témoin de Jéhovah)", "Croyant", 
    "Croyant en tout", "EPC", "Laïc", "Musulman", "Non précisé", "Non-croyant", 
    "Pentecôtiste", "Presbytérien", "Traditionaliste", "UEBC"
]

# Dictionnaire de regroupement
religion_groups = {
    "Christianisme": [
        "Adventiste", "Baptiste", "CMCI", "Chrétien", "Chrétien (Catholique)", 
        "Chrétien (Non précisé)", "Chrétien (Né de nouveau)", "Chrétien (Protestant)", 
        "Chrétien (Témoin de Jéhovah)", "EPC", "Pentecôtiste", "Presbytérien", "UEBC"
    ],
    "Islam": ["Musulman"],
    "Croyances traditionnelles": ["Animiste", "Traditionaliste"],
    "Non religieux": ["Aucune", "Non-croyant", "Non précisé", "Laïc"],
    "Croyances générales": ["Croyant", "Croyant en tout"]
}

# Fonction pour assigner un groupe à une religion
def get_religion_group(religion):
    # Convertir en chaîne et gérer les NaN ou valeurs invalides
    religion = str(religion).strip()
    for group, religions_list in religion_groups.items():
        if religion in religions_list:
            return group
    return "Non précisé"  # Par défaut pour les valeurs inconnues ou non listées

# Corriger uniquement le champ 'religion'
df_volontaire['religion'] = df_volontaire['religion'].apply(get_religion_group)

# Sauvegarder le DataFrame corrigé
df_volontaire.to_csv("df_volontaire_clean_corrige.csv", index=False)

# Afficher un aperçu pour vérification
print("Aperçu du DataFrame corrigé :")
print(df_volontaire[['religion']].head())  # Affiche seulement la colonne 'religion' pour vérification
print("\nRépartition des groupes de religion :")
print(df_volontaire['religion'].value_counts())
```
