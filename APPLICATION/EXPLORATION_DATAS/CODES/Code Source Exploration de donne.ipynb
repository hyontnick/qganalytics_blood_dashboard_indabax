## Visualisation initial

```python
# 1. R√©partition des donneurs par genre
def plot_gender_distribution(df, sheet_name):
    """Visualise la r√©partition des donneurs par genre."""
    gender_col = 'Genre' if 'Genre' in df.columns else 'Genre_' if 'Genre_' in df.columns else 'Sexe'
    gender_counts = df[gender_col].value_counts().reset_index()
    gender_counts.columns = ['Genre', 'Nombre']
    
    fig = px.bar(gender_counts, x='Genre', y='Nombre', 
                 title=f'R√©partition par Genre - {sheet_name}',
                 labels={'Nombre': 'Nombre de Donneurs'})
    fig.show()

# Visualiser pour chaque feuille
plot_gender_distribution(df_2019, '2019')
plot_gender_distribution(df_2020, '2020')
plot_gender_distribution(df_volontaire, 'Volontaire')

# 2. Distribution des √¢ges
def plot_age_distribution(df, sheet_name):
    """Visualise la distribution des √¢ges."""
    age_col = 'Age' if 'Age' in df.columns else 'age_calcul√©'  # Nous calculerons l‚Äô√¢ge pour "2019" plus tard
    if age_col in df.columns:
        fig = px.histogram(df, x=age_col, nbins=20, 
                           title=f'Distribution des √Çges - {sheet_name}',
                           labels={'x': '√Çge', 'y': 'Nombre de Donneurs'})
        fig.show()
    else:
        print(f"Colonne d‚Äô√¢ge non disponible pour {sheet_name}")

# Visualiser pour "2020" et "Volontaire" (√¢ge disponible)
plot_age_distribution(df_2020, '2020')
plot_age_distribution(df_volontaire, 'Volontaire')

# Pour "2019", calculer l‚Äô√¢ge √† partir de "Date de naissance" d‚Äôabord
df_2019['age_calcul√©'] = (2019 - pd.to_datetime(df_2019['Date de naissance'], errors='coerce').dt.year)
plot_age_distribution(df_2019, '2019')

# 3. Nombre de donneurs par arrondissement
def plot_arrondissement_distribution(df, sheet_name):
    """Visualise le nombre de donneurs par arrondissement."""
    arr_col = 'Arrondissement de r√©sidence' if 'Arrondissement de r√©sidence' in df.columns else 'Arrondissement_de_r√©sidence_'
    if arr_col in df.columns:
        arr_counts = df[arr_col].value_counts().reset_index()
        arr_counts.columns = ['Arrondissement', 'Nombre']
        
        fig = px.bar(arr_counts, x='Arrondissement', y='Nombre', 
                     title=f'R√©partition par Arrondissement - {sheet_name}',
                     labels={'Nombre': 'Nombre de Donneurs'})
        fig.update_layout(xaxis={'tickangle': 45})
        fig.show()
    else:
        print(f"Colonne d‚Äôarrondissement non disponible pour {sheet_name}")

# Visualiser pour "2019" et "Volontaire" (pas d‚Äôarrondissement dans "2020")
plot_arrondissement_distribution(df_2019, '2019')
plot_arrondissement_distribution(df_volontaire, 'Volontaire')
```

## Nettoyage

```python
# Fonction pour standardiser les noms de colonnes
def standardize_columns(df):
    """Standardise les noms de colonnes en supprimant espaces, accents et caract√®res sp√©ciaux."""
    df.columns = (df.columns.str.strip()
                  .str.lower()
                  .str.replace(' ', '_')
                  .str.replace('(', '')
                  .str.replace(')', '')
                  .str.replace(',', '_')
                  .str.replace('/', '_')
                  .str.replace('√©', 'e')
                  .str.replace("'", '')
                  .str.replace('__', '_')
                  .str.rstrip('_'))
    return df
# Liste des colonnes format√©es
volontaire_columns = [
    'id',
    'age',
    'horodateur',
    'niveau_detude',
    'genre',
    'taille',
    'poids',
    'situation_matrimoniale_sm',
    'profession',
    'arrondissement_de_residence',
    'quartier_de_residence',
    'nationalite',
    'religion',
    'a_til_elle_deja_donne_le_sang',
    'si_oui_preciser_la_date_du_dernier_don',
    'taux_dhemoglobine',
    'eligibilite_au_don',
    'raison_indisponibilite_est_sous_anti_biotherapie',
    'raison_indisponibilite_taux_dhemoglobine_bas',
    'raison_indisponibilite_date_de_dernier_don_inf_3_mois',
    'raison_indisponibilite_ist_recente_exclu_vih_hbs_hcv',
    'date_de_dernieres_regles_ddr',
    'raison_de_lindisponibilite_de_la_femme_la_ddr_est_mauvais_si_inf_14_jour_avant_le_don',
    'raison_de_lindisponibilite_de_la_femme_allaitement',
    'raison_de_lindisponibilite_de_la_femme_a_accoucher_ces_6_derniers_mois',
    'raison_de_lindisponibilite_de_la_femme_interruption_de_grossesse_ces_06_derniers_mois',
    'raison_de_lindisponibilite_de_la_femme_est_enceinte',
    'autre_raisons_preciser',
    'selectionner_ok_pour_envoyer',
    'raison_de_non_eligibilite_totale_antecedent_de_transfusion',
    'raison_de_non_eligibilite_totale_porteur_hiv_hbs_hcv',
    'raison_de_non_eligibilite_totale_opere',
    'raison_de_non_eligibilite_totale_drepanocytaire',
    'raison_de_non_eligibilite_totale_diabetique',
    'raison_de_non_eligibilite_totale_hypertendus',
    'raison_de_non_eligibilite_totale_asthmatiques',
    'raison_de_non_eligibilite_totale_cardiaque',
    'raison_de_non_eligibilite_totale_tatoue',
    'raison_de_non_eligibilite_totale_scarifie',
    'si_autres_raison_preciser'
]

# Charger df_volontaire et appliquer les nouveaux noms
# Assurez-vous que 'dataset.xlsx' est le bon chemin
df_volontaire = pd.read_excel('Challenge dataset.xlsx', sheet_name='Volontaire')
df_volontaire.columns = volontaire_columns  # Remplacer les noms originaux

# Fonction pour nettoyer df_volontaire avec horodateur ajust√©
def clean_volontaire(df):
    """Nettoie 'Volontaire' et supprime la colonne 'horodateur'."""
    df.columns = volontaire_columns  # Appliquer les noms format√©s
    df = df.copy()
    
    # Supprimer la colonne 'horodateur'
    df.drop(columns=['horodateur'], inplace=True)
    
    # Convertir les dates
    df['si_oui_preciser_la_date_du_dernier_don'] = pd.to_datetime(df['si_oui_preciser_la_date_du_dernier_don'], errors='coerce')
    df['date_de_dernieres_regles_ddr'] = pd.to_datetime(df['date_de_dernieres_regles_ddr'], errors='coerce')
    df['taux_dhemoglobine'] = pd.to_numeric(df['taux_dhemoglobine'], errors='coerce')
    
    # G√©rer les valeurs manquantes
    for col in ['taille', 'poids', 'taux_dhemoglobine']:
        df[col].fillna(df[col].mean(), inplace=True)
    
    categorical_cols = ['niveau_detude', 'genre', 'situation_matrimoniale_sm', 'profession',
                        'arrondissement_de_residence', 'quartier_de_residence', 'nationalite', 'religion']
    for col in categorical_cols:
        df[col].fillna(df[col].mode()[0], inplace=True)
    
    bool_cols = [col for col in df.columns if 'raison_' in col]
    for col in bool_cols:
        df[col].fillna('Inconnu', inplace=True)
    
    df['autre_raisons_preciser'].fillna('Aucune', inplace=True)
    df['si_autres_raison_preciser'].fillna('Aucune', inplace=True)
    
    df.drop_duplicates(subset=['id'], keep='first', inplace=True)
    
    return df

# Nettoyer df_2019 et extraire les dates de 2019
def clean_and_extract_2019_dates(df):
    """Nettoie '2019' et extrait 'date_de_remplissage_de_la_fiche' pour 2019 uniquement."""
    df = standardize_columns(df.copy())
    
    # Convertir 'date_de_remplissage_de_la_fiche' au format JJ/MM/AA
    df['date_de_remplissage_de_la_fiche'] = pd.to_datetime(df['date_de_remplissage_de_la_fiche'], 
                                                          format='%d/%m/%y', errors='coerce')
    
    # Filtrer pour ne garder que les dates de 2019
    df_2019_dates = df[['date_de_remplissage_de_la_fiche']].copy()
    df_2019_dates = df_2019_dates[df_2019_dates['date_de_remplissage_de_la_fiche'].dt.year == 2019]
    
    return df_2019_dates

# Fonction pour nettoyer df_2020
def clean_2020(df):
    """Nettoie les donn√©es de la feuille '2020'."""
    df = standardize_columns(df.copy())
    
    # Convertir 'horodateur' en datetime (d√©j√† correct, mais on s‚Äôassure)
    df['horodateur'] = pd.to_datetime(df['horodateur'], errors='coerce')
    
    # Convertir 'age' en int (actuellement object)
    df['age'] = pd.to_numeric(df['age'], errors='coerce').astype('Int64')  # Int64 g√®re les NaN
    
    # G√©rer les valeurs manquantes
    # 'type_de_donation' : imputer avec la mode
    df['type_de_donation'].fillna('I', inplace=True)
    
    # V√©rifier les doublons (pas d‚ÄôID, utiliser 'horodateur' comme proxy)
    df.drop_duplicates(subset=['horodateur'], keep='first', inplace=True)
    
    return df

# Appliquer le nettoyage
df_volontaire_clean = clean_volontaire(df_volontaire)
df_2020_clean = clean_2020(df_2020)
df_2019_dates = clean_and_extract_2019_dates(df_2019)

# V√©rifier les r√©sultats
print("Aper√ßu de df_volontaire apr√®s nettoyage :")
display(df_volontaire_clean.head())
print("\nInfo de df_volontaire apr√®s nettoyage :")
df_volontaire_clean.info()

print("\nAper√ßu de df_2020 apr√®s nettoyage :")
display(df_2020_clean.head())
print("\nInfo de df_2020 apr√®s nettoyage :")
df_2020_clean.info()

print("\nAper√ßu des dates extraites de 2019 :")
display(df_2019_dates.head())
print("\nInfo des dates extraites de 2019 :")
df_2019_dates.info()

# Sauvegarder les donn√©es nettoy√©es
df_volontaire_clean.to_csv('volontaire_clean.csv', index=False)
df_2020_clean.to_csv('2020_clean.csv', index=False)
df_2019_dates.to_csv('dates_2019_extraites.csv', index=False)
print("\nDonn√©es nettoy√©es sauvegard√©es dans 'volontaire_clean.csv' ,'2020_clean.csv' et 'dates_2019_extraites.csv'")
```

```python
# Importer les biblioth√®ques n√©cessaires
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import mutual_info_classif, RFE
from sklearn.svm import LinearSVC
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report
import xgboost as xgb
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

# Charger le dataframe
df_volontaire = pd.read_csv('volontaire_clean_corrige.csv')

# Pr√©paration des donn√©es
features = [
    'age', 'niveau_detude', 'genre', 'taille', 'poids', 'situation_matrimoniale_sm', 
    'profession', 'arrondissement_de_residence', 'nationalite', 'religion', 
    'a_til_elle_deja_donne_le_sang', 'si_oui_preciser_la_date_du_dernier_don', 
    'taux_dhemoglobine', 'raison_indisponibilite_est_sous_anti_biotherapie', 
    'raison_indisponibilite_taux_dhemoglobine_bas', 'raison_indisponibilite_date_de_dernier_don_inf_3_mois', 
    'raison_indisponibilite_ist_recente_exclu_vih_hbs_hcv', 'date_de_dernieres_regles_ddr', 
    'raison_de_lindisponibilite_de_la_femme_la_ddr_est_mauvais_si_inf_14_jour_avant_le_don', 
    'raison_de_lindisponibilite_de_la_femme_allaitement', 
    'raison_de_lindisponibilite_de_la_femme_a_accoucher_ces_6_derniers_mois', 
    'raison_de_lindisponibilite_de_la_femme_interruption_de_grossesse_ces_06_derniers_mois', 
    'raison_de_lindisponibilite_de_la_femme_est_enceinte', 
    'raison_de_non_eligibilite_totale_antecedent_de_transfusion', 
    'raison_de_non_eligibilite_totale_porteur_hiv_hbs_hcv', 
    'raison_de_non_eligibilite_totale_opere', 'raison_de_non_eligibilite_totale_drepanocytaire', 
    'raison_de_non_eligibilite_totale_diabetique', 'raison_de_non_eligibilite_totale_hypertendus', 
    'raison_de_non_eligibilite_totale_asthmatiques', 'raison_de_non_eligibilite_totale_cardiaque', 
    'raison_de_non_eligibilite_totale_tatoue', 'raison_de_non_eligibilite_totale_scarifie'
]
target = 'eligibilite_au_don'

X = df_volontaire[features].copy()
# Correction : Mapper les trois valeurs explicitement
y = df_volontaire[target].map({
    'Eligible': 1, 
    'D√©finitivement non-eligible': 0, 
    'Temporairement Non-eligible': 0
})

# V√©rifier s‚Äôil reste des NaN dans y
if y.isna().sum() > 0:
    print(f"Attention : {y.isna().sum()} valeurs NaN dans y. Suppression des lignes concern√©es.")
    valid_idx = y.notna()
    X = X[valid_idx]
    y = y[valid_idx]

# Encoder les variables cat√©goriques
le = LabelEncoder()
for col in X.select_dtypes(include=['object']).columns:
    X[col] = le.fit_transform(X[col].astype(str))

# Convertir les dates en jours
ref_date = pd.Timestamp('1900-01-01')
X['si_oui_preciser_la_date_du_dernier_don'] = (pd.to_datetime(X['si_oui_preciser_la_date_du_dernier_don'], errors='coerce') - ref_date).dt.days
X['date_de_dernieres_regles_ddr'] = (pd.to_datetime(X['date_de_dernieres_regles_ddr'], errors='coerce') - ref_date).dt.days
X.fillna(-1, inplace=True)

# Standardiser les donn√©es
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Diviser en ensembles d‚Äôentra√Ænement et de test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# --- 1. Information Gain (Mutual Information) ---
mi_scores = mutual_info_classif(X_train, y_train, random_state=42)
mi_df = pd.DataFrame({'Feature': features, 'MI_Score': mi_scores}).sort_values('MI_Score', ascending=False)
print("\nTop 10 Features - Information Gain:")
print(mi_df.head(10))

# --- 2. Random Forest ---
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
rf_importance = pd.DataFrame({'Feature': features, 'Importance': rf.feature_importances_}).sort_values('Importance', ascending=False)
print("\nTop 10 Features - Random Forest:")
print(rf_importance.head(10))

# --- 3. Logistic Regression ---
lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X_train, y_train)
lr_importance = pd.DataFrame({'Feature': features, 'Coefficient': np.abs(lr.coef_[0])}).sort_values('Coefficient', ascending=False)
print("\nTop 10 Features - Logistic Regression:")
print(lr_importance.head(10))

# --- 4. Recursive Feature Elimination (RFE) avec Logistic Regression ---
rfe = RFE(estimator=LogisticRegression(max_iter=1000), n_features_to_select=10)
rfe.fit(X_train, y_train)
rfe_ranking = pd.DataFrame({'Feature': features, 'Ranking': rfe.ranking_}).sort_values('Ranking')
print("\nTop 10 Features - RFE (Logistic Regression):")
print(rfe_ranking.head(10))

# --- 5. XGBoost ---
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb_model.fit(X_train, y_train)
xgb_importance = pd.DataFrame({'Feature': features, 'Importance': xgb_model.feature_importances_}).sort_values('Importance', ascending=False)
print("\nTop 10 Features - XGBoost:")
print(xgb_importance.head(10))

# --- 6. SVM (Linear Kernel) ---
svm = LinearSVC(max_iter=10000, random_state=42)
svm.fit(X_train, y_train)
svm_importance = pd.DataFrame({'Feature': features, 'Coefficient': np.abs(svm.coef_[0])}).sort_values('Coefficient', ascending=False)
print("\nTop 10 Features - SVM (Linear):")
print(svm_importance.head(10))

# Visualisation des importances (Random Forest comme exemple)
plt.figure(figsize=(12, 6))
sns.barplot(x='Importance', y='Feature', data=rf_importance.head(10))
plt.title('Top 10 Features - Random Forest')
plt.show()

# √âvaluation des mod√®les
models = {
    'Random Forest': rf,
    'Logistic Regression': lr,
    'XGBoost': xgb_model,
    'SVM': svm
}

print("\nPerformances des mod√®les :")
for name, model in models.items():
    y_pred = model.predict(X_test)
    print(f"\n{name} - Rapport de classification :")
    print(classification_report(y_test, y_pred))

# Sauvegarder le meilleur mod√®le (Random Forest par d√©faut)
joblib.dump(rf, 'eligibility_model_rf.pkl')
joblib.dump(scaler, 'scaler.pkl')
print("\nMod√®le Random Forest sauvegard√© dans 'eligibility_model_rf.pkl'")
print("Scaler sauvegard√© dans 'scaler.pkl'")
```

```python
import pandas as pd
import joblib
from datetime import datetime

# Charger le mod√®le et le scaler
model = joblib.load('eligibility_model_rf_no_leak.pkl')
scaler = joblib.load('scaler_no_leak.pkl')

# Liste des features utilis√©es dans le mod√®le
FEATURES = [
    'age', 'niveau_detude', 'genre', 'taille', 'poids', 'situation_matrimoniale_sm',
    'profession', 'arrondissement_de_residence', 'nationalite', 'religion',
    'a_til_elle_deja_donne_le_sang', 'si_oui_preciser_la_date_du_dernier_don',
    'taux_dhemoglobine'
]

# Dictionnaire pour encoder les variables cat√©goriques (bas√© sur ton dataset)
mappings = {
    'niveau_detude': {'Aucun': 0, 'Primaire': 1, 'Secondaire': 2, 'Universitaire': 3, 'Pas Pr√©cis√©': 4},
    'genre': {'Homme': 0, 'Femme': 1},
    'situation_matrimoniale_sm': {'C√©libataire': 0, 'Mari√© (e)': 1},
    'profession': {'Etudiant (e)': 0, 'Sans Emplois': 1, 'Militaire': 2, 'Commer√ßant (e)': 3, 'Tailleur': 4, 'Bijoutier': 5, 'Couturier': 6, 'Jeune cadre': 7},
    'arrondissement_de_residence': {'Douala 1': 0, 'Douala 2': 1, 'Douala 3': 2, 'Douala 5': 3, 'R A S': 4, 'Douala (Non pr√©cis√© )': 5, 'Non pr√©cis√©': 6},
    'nationalite': {'Camerounaise': 0, 'R A S': 1},
    'religion': {'Chretien (Catholique)': 0, 'Musulman': 1, 'Adventiste': 2, 'Pas Pr√©cis√©': 3},
    'a_til_elle_deja_donne_le_sang': {'Oui': 1, 'Non': 0}
}

# Fonction pour collecter les donn√©es utilisateur
def collect_input():
    data = {}
    print("Entrez les informations du donneur :")
    data['age'] = int(input("√Çge : "))
    data['niveau_detude'] = input("Niveau d‚Äô√©tude (ex. Universitaire, Primaire, Aucun) : ")
    data['genre'] = input("Genre (Homme/Femme) : ")
    data['taille'] = float(input("Taille (en cm, ex. 172.8) : "))
    data['poids'] = float(input("Poids (en kg, ex. 77.2) : "))
    data['situation_matrimoniale_sm'] = input("Situation matrimoniale (C√©libataire/Mari√© (e)) : ")
    data['profession'] = input("Profession (ex. Etudiant (e), Sans Emplois, Militaire) : ")
    data['arrondissement_de_residence'] = input("Arrondissement de r√©sidence (ex. Douala 3, R A S) : ")
    data['nationalite'] = input("Nationalit√© (ex. Camerounaise) : ")
    data['religion'] = input("Religion (ex. Chretien (Catholique), Musulman) : ")
    data['a_til_elle_deja_donne_le_sang'] = input("A-t-il/elle d√©j√† donn√© le sang ? (Oui/Non) : ")
    data['si_oui_preciser_la_date_du_dernier_don'] = input("Si oui, date du dernier don (YYYY-MM-DD, sinon 1900-01-01) : ")
    data['taux_dhemoglobine'] = float(input("Taux d‚Äôh√©moglobine (ex. 13.9) : "))
    return data

# Fonction pour encoder les donn√©es
def encode_input(data):
    encoded = data.copy()
    for feature, mapping in mappings.items():
        encoded[feature] = mapping.get(data[feature], -1)  # -1 pour valeurs inconnues
    
    # Convertir la date en jours depuis 1900-01-01
    ref_date = pd.Timestamp('1900-01-01')
    date = pd.to_datetime(data['si_oui_preciser_la_date_du_dernier_don'], errors='coerce')
    encoded['si_oui_preciser_la_date_du_dernier_don'] = (date - ref_date).days if not pd.isna(date) else -1
    
    return encoded

# Faire la pr√©diction
def predict_eligibility():
    # Collecter les donn√©es
    input_data = collect_input()
    
    # Encoder les donn√©es
    encoded_data = encode_input(input_data)
    
    # Cr√©er un dataframe avec les features dans l‚Äôordre attendu
    input_df = pd.DataFrame([encoded_data], columns=FEATURES)
    
    # Standardiser les donn√©es
    input_scaled = scaler.transform(input_df)
    
    # Faire la pr√©diction
    prediction = model.predict(input_scaled)[0]
    probability = model.predict_proba(input_scaled)[0]
    
    # Afficher le r√©sultat
    result = "√âligible" if prediction == 1 else "Non √©ligible"
    print(f"\nR√©sultat de la pr√©diction : {result}")
    print(f"Probabilit√© d‚Äô√™tre √©ligible : {probability[1]:.2%}")
    print(f"Probabilit√© d‚Äô√™tre non √©ligible : {probability[0]:.2%}")

# Ex√©cuter le script
if __name__ == "__main__":
    predict_eligibility()
```

## Optimisation du model

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, roc_auc_score
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import GridSearchCV
import joblib

# Charger le dataframe
df_volontaire = pd.read_csv('df_volontaire_clean_corrige.csv')

# Feature engineering : ajout de l'IMC
def calculate_imc(row):
    return row['poids'] / ((row['taille']/100) ** 2) if row['taille'] > 0 else np.nan

df_volontaire['imc'] = df_volontaire.apply(calculate_imc, axis=1)

# Features align√©es avec le code initial + IMC
features = [
    'age', 'niveau_detude', 'genre', 'taille', 'poids', 'situation_matrimoniale_sm',
    'profession', 'arrondissement_de_residence', 'nationalite', 'religion',
    'a_til_elle_deja_donne_le_sang', 'si_oui_preciser_la_date_du_dernier_don',
    'taux_dhemoglobine', 'imc'
]
target = 'eligibilite_au_don'

X = df_volontaire[features].copy()
y = df_volontaire[target].map({
    'Eligible': 1,
    'D√©finitivement non-eligible': 0,
    'Temporairement Non-eligible': 0
})

# Nettoyage initial
valid_idx = y.notna() & X['taux_dhemoglobine'].notna()
X = X[valid_idx]
y = y[valid_idx]

# Encodage
le_dict = {}
for col in X.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col].astype(str))
    le_dict[col] = le

# Gestion des dates
ref_date = pd.Timestamp('2025-03-18')
X['jours_depuis_dernier_don'] = (pd.to_datetime(X['si_oui_preciser_la_date_du_dernier_don'], 
                                              errors='coerce') - ref_date).dt.days
X = X.drop('si_oui_preciser_la_date_du_dernier_don', axis=1)
X.fillna(-1, inplace=True)

# Standardisation
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Gestion du d√©s√©quilibre des classes avec SMOTE
smote = SMOTE(random_state=42)
X_balanced, y_balanced = smote.fit_resample(X_scaled, y)

# Split
X_train, X_test, y_train, y_test = train_test_split(X_balanced, y_balanced, 
                                                    test_size=0.2, 
                                                    random_state=42, 
                                                    stratify=y_balanced)

# Optimisation des hyperparam√®tres
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5],
    'class_weight': ['balanced', None]
}
rf = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(rf, param_grid, cv=5, scoring='f1_macro', n_jobs=-1)
grid_search.fit(X_train, y_train)

# Meilleur mod√®le
best_rf = grid_search.best_estimator_
print(f"Meilleurs param√®tres: {grid_search.best_params_}")

# Importance des features
feature_importance = pd.DataFrame({
    'Feature': [f for f in features if f != 'si_oui_preciser_la_date_du_dernier_don'] + ['jours_depuis_dernier_don'],
    'Importance': best_rf.feature_importances_
}).sort_values('Importance', ascending=False)
print("\nTop Features - Random Forest optimis√©:")
print(feature_importance)

# √âvaluation
y_pred = best_rf.predict(X_test)
y_pred_proba = best_rf.predict_proba(X_test)[:, 1]
print("\nRapport de classification optimis√©:")
print(classification_report(y_test, y_pred))
print(f"AUC-ROC Score: {roc_auc_score(y_test, y_pred_proba):.3f}")

# Validation crois√©e
cv_scores = cross_val_score(best_rf, X_balanced, y_balanced, cv=5, scoring='f1_macro')
print(f"\nCross-validation F1-macro scores: {cv_scores}")
print(f"Moyenne CV F1-macro: {cv_scores.mean():.3f} (+/- {cv_scores.std() * 2:.3f})")

# Sauvegarde
joblib.dump(best_rf, 'eligibility_model_rf_optimized.pkl')
joblib.dump(scaler, 'scaler_optimized.pkl')
joblib.dump(le_dict, 'label_encoders.pkl')
print("\nMod√®le sauvegard√© dans 'eligibility_model_rf_optimized.pkl'")
print("Scaler sauvegard√© dans 'scaler_optimized.pkl'")
print("Encodeurs sauvegard√©s dans 'label_encoders.pkl'")
```

```python
import pandas as pd
import joblib
from datetime import datetime

# Charger le mod√®le, le scaler et les encodeurs
model = joblib.load('eligibility_model_rf_optimized.pkl')
scaler = joblib.load('scaler_optimized.pkl')
le_dict = joblib.load('label_encoders.pkl')

# R√©cup√©rer les noms des features attendus par le scaler
expected_features = scaler.feature_names_in_ if hasattr(scaler, 'feature_names_in_') else None
if expected_features is None:
    print("Impossible de r√©cup√©rer les features attendues automatiquement. Utilisation de l'ordre suppos√©.")
    FEATURES = [
        'age', 'niveau_detude', 'genre', 'taille', 'poids', 'situation_matrimoniale_sm',
        'profession', 'arrondissement_de_residence', 'nationalite', 'religion',
        'a_til_elle_deja_donne_le_sang', 'jours_depuis_dernier_don', 'taux_dhemoglobine', 'imc'
    ]
else:
    print("Features attendues par le mod√®le :", list(expected_features))
    FEATURES = list(expected_features)

# Fonction pour calculer l'IMC
def calculate_imc(taille, poids):
    return poids / ((taille/100) ** 2) if taille > 0 else -1

# Fonction pour collecter les donn√©es utilisateur
def collect_input():
    data = {}
    print("Entrez les informations du donneur :")
    data['age'] = int(input("√Çge : "))
    data['niveau_detude'] = input("Niveau d‚Äô√©tude (ex. Universitaire, Primaire, Aucun) : ")
    data['genre'] = input("Genre (Homme/Femme) : ")
    data['taille'] = float(input("Taille (en cm, ex. 172.8) : "))
    data['poids'] = float(input("Poids (en kg, ex. 77.2) : "))
    data['situation_matrimoniale_sm'] = input("Situation matrimoniale (C√©libataire/Mari√© (e)) : ")
    data['profession'] = input("Profession (ex. Etudiant, Sans emploi, Militaire) : ")
    data['arrondissement_de_residence'] = input("Arrondissement de r√©sidence (ex. Douala 3, R A S) : ")
    data['nationalite'] = input("Nationalit√© (ex. Camerounaise) : ")
    data['religion'] = input("Religion (ex. Chretien (Catholique), Musulman) : ")
    data['a_til_elle_deja_donne_le_sang'] = input("A-t-il/elle d√©j√† donn√© le sang ? (Oui/Non) : ")
    data['si_oui_preciser_la_date_du_dernier_don'] = input("Si oui, date du dernier don (YYYY-MM-DD, sinon laissez vide) : ")
    data['taux_dhemoglobine'] = float(input("Taux d‚Äôh√©moglobine (ex. 13.9) : "))
    return data

# Fonction pour encoder les donn√©es
def encode_input(data):
    encoded = {}
    
    # Encodage selon l'ordre des FEATURES
    for feature in FEATURES:
        if feature == 'age':
            encoded[feature] = data['age']
        elif feature == 'taille':
            encoded[feature] = data['taille']
        elif feature == 'poids':
            encoded[feature] = data['poids']
        elif feature == 'taux_dhemoglobine':
            encoded[feature] = data['taux_dhemoglobine']
        elif feature == 'imc':
            encoded[feature] = calculate_imc(data['taille'], data['poids'])
        elif feature == 'jours_depuis_dernier_don':
            ref_date = pd.Timestamp('2025-03-18')
            date_input = data['si_oui_preciser_la_date_du_dernier_don']
            if date_input.strip() and data['a_til_elle_deja_donne_le_sang'].lower() == 'oui':
                date = pd.to_datetime(date_input, errors='coerce')
                encoded[feature] = (date - ref_date).days if not pd.isna(date) else -1
            else:
                encoded[feature] = -1
        else:
            # Features cat√©goriques
            try:
                encoded[feature] = le_dict[feature].transform([str(data[feature])])[0]
            except (KeyError, ValueError):
                encoded[feature] = -1
    
    return encoded

# Faire la pr√©diction
def predict_eligibility():
    # Collecter les donn√©es
    input_data = collect_input()
    
    # Encoder les donn√©es
    encoded_data = encode_input(input_data)
    
    # Cr√©er un dataframe avec les features dans l‚Äôordre exact
    input_df = pd.DataFrame([encoded_data], columns=FEATURES)
    
    # V√©rification de l'ordre des colonnes
    print("Ordre des colonnes dans input_df :", list(input_df.columns))
    
    # Remplacer les NaN par -1
    input_df.fillna(-1, inplace=True)
    
    # Standardiser les donn√©es
    input_scaled = scaler.transform(input_df)
    
    # Faire la pr√©diction
    prediction = model.predict(input_scaled)[0]
    probability = model.predict_proba(input_scaled)[0]
    
    # Afficher le r√©sultat
    result = "√âligible" if prediction == 1 else "Non √©ligible"
    print(f"\nR√©sultat de la pr√©diction : {result}")
    print(f"Probabilit√© d‚Äô√™tre √©ligible : {probability[1]:.2%}")
    print(f"Probabilit√© d‚Äô√™tre non √©ligible : {probability[0]:.2%}")

# Ex√©cuter le script
if __name__ == "__main__":
    predict_eligibility()
```
